### Select the feature and label that will be used for Regression modeling

```python
cleaned_datast = dataset[['ENGINE SIZE (L)','CO2_EMISSIONS']]
```

```python
cleaned_datast.head()
```

### Split the dataset to train and test

```python
msk = np.random.rand(len(cleaned_datast)) < 0.8
train = cleaned_datast[msk]
test = cleaned_datast[~msk]
print("Train dataset shape" , train.shape)
print("Test dataset shape" , test.shape)
```

```python
train_x = np.asanyarray(train[['ENGINE SIZE (L)']])
train_y = np.asanyarray(train[['CO2_EMISSIONS']])
```

```python
test_x = np.asanyarray(test[['ENGINE SIZE (L)']])
test_y = np.asanyarray(test[['CO2_EMISSIONS']])
```

# Modeling

### Linear Regression Modeling with the engine size the feature and CO2_EMISSIONS as a label

### Import the required library

```python
from sklearn import linear_model
```

### Start modeling and train the model

```python
regr = linear_model.LinearRegression()
regr.fit (train_x, train_y)
# The coefficients
print ('Coefficients: ', regr.coef_)
print ('Intercept: ',regr.intercept_)
```

### Plot the model

```python
plt.scatter(train['ENGINE SIZE (L)'], train['CO2_EMISSIONS'],  color='blue')
plt.plot(train_x, regr.coef_[0][0]*train_x + regr.intercept_[0], '-r')
plt.xlabel("Engine size")
plt.ylabel("Emission")
```

### Evaluate the model

```python
from sklearn.metrics import r2_score

test_x = np.asanyarray(test[['ENGINE SIZE (L)']])
test_y = np.asanyarray(test[['CO2_EMISSIONS']])
test_y_hat = regr.predict(test_x)

print("Mean absolute error: %.2f" % np.mean(np.absolute(test_y_hat - test_y)))
print("Residual sum of squares (MSE): %.2f" % np.mean((test_y_hat - test_y) ** 2))
print("R2-score: %.2f" % r2_score(test_y_hat , test_y) )
```
